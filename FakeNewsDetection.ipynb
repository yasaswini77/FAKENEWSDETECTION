```python
data_merge = pd.concat([data_fake, data_true],axis = 0)
data_merge.head(10)
```

```python
data = data_merge.drop(['title','subject','date'],axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
# shuffles the rows of the DataFrame data
# frac=1 indicates that it selects all rows, essentially shuffling the entire DataFrame. This operation is commonly used to randomize the order of rows in a dataset.
```

```python
data.head()
```

```python
data.reset_index(inplace=True)
data.drop(['index'],axis=1,inplace=True)

# inplace=True modifies the DataFrame in place, meaning it doesn't return a new DataFrame but instead modifies the existing one.
# adds a new column named "index" containing the old index values and resets the index to the default integer index (0, 1, 2, ...)
# drop() function is used to remove columns or rows from a DataFrame. 
# axis=1 indicates that it's a column operation.
```

```python
data.columns
```

```python
data.head()
```

```python
import re

def wordopt(text):
    if isinstance(text, str):
        text = text.lower()
        text = re.sub('\[.*?\]','',text)
        text = re.sub("\\W"," ",text)
        text = re.sub('https?://\S+|www\.\S+', '', text)
        text = re.sub('<.*?>+', '', text)
        text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
        text = re.sub('\n', '', text)
        text = re.sub('\w*\d\w*', '', text)
    else:
        text = str(text)  # Convert non-string values to empty strings
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x=data['text']
y=data['class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vector = TfidfVectorizer()
tf_train = vector.fit_transform(x_train)
tf_test = vector.transform(x_test)
```

```python
print(tf_train)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(tf_train, y_train)
```

```python
y.pred = LR.predict(tf_test)
```

```python
accuracy= LR.score(tf_test, y_test)
from sklearn.metrics import precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import numpy as np


# Assuming LR is your logistic regression model
# Assuming tf_test is your test data after TF-IDF transformation
# Assuming y_test is your test labels

# Get predictions from your LR model
predictions = LR.predict(tf_test)

# Calculate precision, recall, and F1-score
precision = precision_score(y_test, predictions)
recall = recall_score(y_test, predictions)
f1 = f1_score(y_test, predictions)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)

labels = ['Accuracy', 'Precision', 'F1 Score', 'Recall']

# Scores
scores = [accuracy, precision, f1, recall]

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(labels, scores, color=['blue', 'green', 'orange', 'red'])
plt.title('Performance Metrics for Logistic Regression')
plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.ylim(0, 1)  # Limit y-axis from 0 to 1
plt.show()
```
